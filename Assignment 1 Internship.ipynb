# Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
def Top100Movie(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    year=soup.find_all('span',class_='secondaryInfo')
    Year_of_Release= []
    for i in year:
        Year_of_Release.append(i.get_text())
    rating=soup.find_all('td',class_='ratingColumn imdbRating')
    IMDB_rating= []
    for i in rating:
        IMDB_rating.append(i.get_text().replace('\n',' '))
    name=soup.find_all('td',class_='titleColumn')
    Movie_name=[]
    for i in name:
        Movie_name.append(i.get_text().replace('\n',''))
    Movies=pd.DataFrame({})
    Movies['Title'] = Movie_name
    Movies['Year_of_Release'] = Year_of_Release
    Movies['IMDB_rating'] = IMDB_rating
    Top100 = Movies.iloc[0:100,:]
    Top100.to_csv(r'C:\Users\Mayur\topmovies.csv')
    return Top100
```

```python
Top100Movie('https://www.imdb.com/chart/top?sort=rk,asc&mode=simple&page=1')
```

# Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
def Book(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    name=soup.find_all('h4',class_='italic')
    Book_name=[]
    for i in name:
        Book_name.append(i.text.replace('\n',''))
    author=soup.find_all('p',class_='sans bold')
    Authorname=[]
    for i in author:
        Authorname.append(i.text.replace('\n',''))
    gen=soup.find_all('p',class_='genre-links hidden-phone')
    Genre=[]
    for i in gen:
        Genre.append(i.text.replace('\n',''))
    review=soup.find_all("p",class_="excerpt")
    Bookreview=[]
    for i in review:
        Bookreview.append(i.text.replace('\n',''))
    Bookpage=pd.DataFrame({})
    Bookpage['Book name'] = Book_name
    Bookpage['Author'] = Authorname
    Bookpage['Genre'] = Genre
    Bookpage['Review'] = Bookreview
    Bookpage.to_csv('Book review.csv')
    n=5
    Bookpage=Bookpage.sample(5)
    
    return Bookpage
```

```python
Book('https://bookpage.com/reviews/')
```

# Top 10 ODI teams in men's cricket along with the records for matches, points and rating

```python
def Top10(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    team=soup.find_all('span',class_='u-hide-phablet')
    Team=[]
    for i in team:
        Team.append(i.get_text().replace('\n',''))
    mat=soup.find_all('td',class_='rankings-block__banner--matches')
    match=[]
    for i in mat:
        match.append(i.get_text().replace('\n',''))
    
    mtch=soup.find_all("td",class_="table-body__cell u-center-text")
    Match=[]
    for  i in mtch:
        Match.append(i.get_text().replace("\n",""))
    Match=match+Match[::2]
    p = soup.find_all('td',class_="rankings-block__banner--points")
    pnt=[]
    for i in p:
        pnt.append(i.get_text().replace('\n',''))
    poi = soup.find_all('td',class_="table-body__cell u-center-text")
    points=[]
    for i in poi:
        points.append(i.get_text().replace('\n',''))
    points = pnt + points[1::2]
    r=[]
    rtn=soup.find_all("td",class_="rankings-block__banner--rating u-text-right")
    for i in rtn:
        r.append(i.get_text().replace("\n",""))

    def ls(a):  
        str1 = " " # initializing an empty str
        x=str1.join(a)
        y=x.split()
        newr=[]
        for i in y:
            newr.append(i)
        return(newr)

    R_new=ls(r)
    rating=soup.find_all("td",class_="table-body__cell u-text-right rating")
    Rating=[]
    for t in rating:
        Rating.append(t.get_text().replace("\n",""))

    Rating=R_new+Rating
    Top10_ODI=pd.DataFrame({})
    Top10_ODI["Team"]=Team[0:10]
    Top10_ODI["Matches"]=Match[0:10]
    Top10_ODI["Points"]=points[0:10]
    Top10_ODI["Rating"]=Rating[0:10]
    return Top10_ODI 
```

```python
Top10('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

# ii) Top 10 ODI Batsmen in men along with the records of their team and rating.

```python
def Top10men(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    n=soup.find_all('td',class_="table-body__cell name")
    Name=[]
    for i in n:
        Name.append(i.get_text().replace('\n',' '))
    Name=Name[0:10]
    team=soup.find_all('span',class_="table-body__logo-text")
    Team=[]
    for i in team:
        Team.append(i.get_text().replace('\n',' '))
    Team=Team[0:10]
    rating=soup.find_all('td',class_="table-body__cell u-text-right rating")
    Rating=[]
    for i in rating:
        Rating.append(i.get_text().replace('\n',' '))
    Rating=Rating[0:10]
    Topmen=pd.DataFrame({})
    Topmen['Name'] = Name
    Topmen['Team'] = Team
    Topmen['Rating'] = Rating
    return Topmen
```

# Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
def weatherforecasting(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    Period=[]
    Short_description=[]
    Temperature=[]
    Description=[]

    period=soup.find_all('p',class_='period-name')
    for i in period:
        Period.append(i.get_text().replace('\n',' '))
    len(Period)
    short_desc=soup.find_all('p',class_='short-desc')
    for i in short_desc:
        Short_description.append(i.get_text().replace('\n',' '))
    len(Short_description)
    temp1=soup.find_all('p',class_="temp temp-low")
    Temp1=[]
    for i in temp1:
        Temp1.append(i.get_text().replace('\n',''))
    temp2=soup.find_all('p',class_="temp temp-high")
    Temp2=[]
    for i in temp2:
        Temp2.append(i.get_text().replace('\n',''))
    Temp2
    Temperature=Temp1 + Temp2
    description=soup.find_all('div',class_="col-sm-10 forecast-text")
    for i in description:
        Description.append(i.get_text().replace('\n',' '))
    Description=Description[0:9]
    Weather=pd.DataFrame({'Period':Period,'Short_description':Short_description,'Temperature':Temperature,'Description':Description})
    return Weather
```
